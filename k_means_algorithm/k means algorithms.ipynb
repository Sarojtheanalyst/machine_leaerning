# apply the k means algorithm to mour datasets

```python
from sklearn.cluster import KMeans
print(" import library sucessfully")
```

```python
# error handled sucessfully 
wcss=[]  # always use clusters from the 2 
for i in range(2,11):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit_predict(df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(2,11),wcss)
```

# now make the clusters basic 

```python
X=df.iloc[:,:].values   # type nd arrays  [[5.13,88],[5.9,113]]
```

```python
km=KMeans(n_clusters=4)

y_means=km.fit_predict(X)
```

```python
np.unique(y_means)
```

```python
# clustering 2 groups
X[y_means==2]
print(X[y_means==2,0])    # clusters group 2 
print("\n")
print(X[y_means==2,1])
```

```python
# plottings
plt.xlabel("cpga")
plt.ylabel('iq')
plt.title("CGPA vs IQ ")
plt.scatter(X[y_means==0,0],X[y_means==0,1],color='Blue')
plt.scatter(X[y_means==1,0],X[y_means==1,1],color='red')
plt.scatter(X[y_means==2,0],X[y_means==2,1],color='green')
plt.scatter(X[y_means==3,0],X[y_means==3,1],color='yellow')
plt.show()
```
