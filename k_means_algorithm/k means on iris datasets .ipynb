# sepal length and sepal width

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
print(" import library sucessfully")
```

```python
df.drop(columns=['Id','SepalLengthCm','SepalWidthCm','Species'],inplace=True)
```

```python
df.head()
```

# min max scalers

```python
# elbow methods 
wcss=[]  # always use clusters from the 2 
for i in range(2,11):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit_predict(df)
    wcss.append(kmeans.inertia_)
print("sucessfully ")
```

```python
# see all the elbow methods
plt.plot(range(2,11),wcss)
```

```python
X=df.iloc[:,:].values
```

```python
km=KMeans(n_clusters=4)

y_means=km.fit_predict(X)  # this is the predicted values 
```

```python
# find the centroids 
km.cluster_centers_
```

```python
km.cluster_centers_[:,0]
```

```python
km.cluster_centers_[:,1]
```

```python
# plot all the clusters 
plt.scatter(X[y_means==0,0],X[y_means==0,1],color='Blue')
plt.scatter(X[y_means==1,0],X[y_means==1,1],color='red')
plt.scatter(X[y_means==2,0],X[y_means==2,1],color='green')
plt.scatter(X[y_means==3,0],X[y_means==3,1],color='yellow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*',s=220)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title("clustering sepal length and sepal width")
plt.show()
```
